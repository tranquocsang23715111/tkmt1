```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('StudentSurvey.csv')
df.head(10)
```

```python
df["Year"].value_counts(dropna= True)
```

```python
#sinh vien deo khuyen
tongSinhVienDeoKhuyen = (df['Piercings'] >=  1).sum()
tiLeSinhVienDeoKhuyen = tongSinhVienDeoKhuyen/df.shape[0]
tiLeSinhVienDeoKhuyen
```

```python
namDeoKhuyen = df[(df['Gender'] == 'M') & (df['Piercings'] >=1)].shape[0]
tongNam = (df['Gender'] =="M").shape[0]

tiLeNamDeoKhuyen = namDeoKhuyen/tongNam


nuDeoKhuyen = df[(df['Gender'] == 'F') & (df['Piercings'] >=1)].shape[0]
tongNu = (df['Gender'] =="F").shape[0]
tiLeNuDeoKhuyen = nuDeoKhuyen/tongNu

if tiLeNamDeoKhuyen>tiLeNuDeoKhuyen:
    print('nam deo khuyen nhieu hon nu')
else:
    print('nu deo nhieu hon nam')
```

```python
df_sam = df.sample(50)
df_sam
```

```python
df_sam_nu = df_sam[df_sam['Gender'] == 'F'].shape[0]
tiLeNu = df_sam_nu/df_sam.shape[0]
tiLeNuBanDau = (df[df["Gender"] =="F"].shape[0])/(df.shape[0])
print('ti le nu tu mau: ',tiLeNu)
print('sai so voi bo du lieu ban dau la: ', abs(tiLeNuBanDau - tiLeNu))
```

```python
tap_mau = []
for x in range(100):
    df_sam = df.sample(50)
    df_sam_nu = df_sam[df_sam['Gender'] == 'F'].shape[0]
    tiLeNu = df_sam_nu/df_sam.shape[0]
    tap_mau.append(tiLeNu)
plt.hist(tap_mau)
plt.show()
```

```python
tap_mau_70 =[]
tap_mau_100 = []
tap_mau_200 = []

for x in range(100):
    df_sam = df.sample(70)
    df_sam_nu = df_sam[df_sam['Gender'] == 'F'].shape[0]
    tiLeNu = df_sam_nu/df_sam.shape[0]
    tap_mau_70.append(tiLeNu)

for x in range(100):
    df_sam = df.sample(100)
    df_sam_nu = df_sam[df_sam['Gender'] == 'F'].shape[0]
    tiLeNu = df_sam_nu/df_sam.shape[0]
    tap_mau_100.append(tiLeNu)

for x in range(100):
    df_sam = df.sample(200)
    df_sam_nu = df_sam[df_sam['Gender'] == 'F'].shape[0]
    tiLeNu = df_sam_nu/df_sam.shape[0]
    tap_mau_200.append(tiLeNu)

plt.figure(figsize=(10,5))
plt.subplot(2,2,1)
plt.hist(tap_mau)
plt.title('mau 50')

plt.subplot(2,2,2)
plt.hist(tap_mau_70)
plt.title('mau 70')

plt.subplot(2,2,3)
plt.hist(tap_mau_100)
plt.title('mau 100')

plt.subplot(2,2,4)
plt.hist(tap_mau_200)
plt.title('mau 200')
plt.show()
```
