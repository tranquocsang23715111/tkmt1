```python
nguoiTapTheDuc = df[df['exerany'] == 1]
nguoiDanhGia_poor_or_fair = nguoiTapTheDuc[(nguoiTapTheDuc['genhlth'] == 'poor')|(nguoiTapTheDuc['genhlth']== 'fair')].count()
tiTe_danhGia_poor_or_farl = nguoiDanhGia_poor_or_fair.shape[0]/nguoiTapTheDuc.shape[0] * 100
print('ti le danh gia poor or fair la ',round(tiTe_danhGia_poor_or_farl,2),'%')
```

```python
df['height'] = df['height']* 2.54
df['weight'] = df['weight'] * 0.453592 
df['wtdesire'] = df['wtdesire'] *0.453592 
df
```

```python

soNguoiMuonGiamCan = df[df['wtdesire'] <= df['weight']]
tiLeSoNguoiMuonGiamCam = soNguoiMuonGiamCan.shape[0]/df.shape[0]
print('tỉ lệ số người muốn giảm cân là: ',(tiLeSoNguoiMuonGiamCam),'%')
```

```python
from scipy import stats
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in numeric_cols:
    stat, p = stats.shapiro(df[col].dropna())
    print(f"{col}: p-value = {p:.4f}")
    if p > 0.05:
        print("  Có thể coi là phân phối chuẩn\n")
    else:
        print("  Không phải phân phối chuẩn\n")
    
```

```python
df['BMI'] = df['weight']/(df['height']/100)**2
df
```

```python
nam = df[df['gender'] == 'm']
nu = df[df['gender'] == 'f']
```

```python
trungBinhNam = nam['BMI'].mean()
trungBinhNu = nu['BMI'].mean()
print('BMI trung bình nam: ',round(trungBinhNam,2))
print('BMI trung bình nữ: ',round(trungBinhNu,2))
```

```python
plt.bar(['nam','nữ'],[trungBinhNam,trungBinhNu])
```

```python
namBeoPhi = nam[nam['BMI'] >= 30]
nuBeoPhi = nu[nu['BMI'] >=30]
```

```python
tiLeNamBeoPhi = namBeoPhi.shape[0]/nam.shape[0]
tiLeNuBeoPhi = nuBeoPhi.shape[0]/nu.shape[0]
print(f'tỉ lệ nam béo phì là: {tiLeNamBeoPhi:.2f} %')
print(f'tỉ lệ nữ béo phì là: {tiLeNuBeoPhi:.2f} %')
```
